## Usos habitacionales

```python
# creación de dataframe sólo para usos habitacionales
dfh = df[df['USO_DESTIN']=='HABITACIONAL']
dfh = dfh.set_index('ID')
```

```python
# Tabla proyectos por clasificación
tabla = dfh.groupby('COD_COMUNA').agg({'SUPERFICIE': ['sum'], 'CANTIDAD_U': ['sum'], 'NUM_PISOS': ['sum', 'count']})
#tabla.sort_values([('SUPERFICIE', 'sum')], ascending=False, inplace=True)
tabla
```

```python
tabla.columns = ["_".join(x) for x in tabla.columns.ravel()] # sacar los multi índices simplifica un montón trabajar con los datos
tabla.columns
```

```python
tabla.sort_values('SUPERFICIE_sum', ascending=False)
```

```python
# Tablas para usos habitacionales
tabla = dfh.groupby('COD_REGION').agg({'SUPERFICIE': ['mean'], 'CANTIDAD_U': ['mean'], 'NUM_PISOS': ['mean', 'count']})
tabla['num_permisos'] = tabla['NUM_PISOS']['count']
tabla.drop([('NUM_PISOS','count')], axis=1, inplace=True)
tabla['sup_promedio'] = tabla['SUPERFICIE']['mean'] / tabla['CANTIDAD_U']['mean']
#tabla.to_csv(os.path.join(output_f, 'clasificacion.csv'))
print(tabla)
```

## Restringe análisis aún más, a la Región Metropolitana

```python
dfhr = dfh[dfh['COD_REGION'] == 13]
dfhr.head()
```

```python
# Tablas para usos habitacionales
tabla = dfhr.groupby('GLOSA_COMU').agg({'SUPERFICIE': ['mean'], 'CANTIDAD_U': ['mean'], 'NUM_PISOS': ['mean', 'count']})
tabla['num_permisos'] = tabla['NUM_PISOS']['count']
tabla.drop([('NUM_PISOS','count')], axis=1, inplace=True)
tabla['sup_promedio'] = tabla['SUPERFICIE']['mean'] / tabla['CANTIDAD_U']['mean']
# tabla.to_csv(os.path.join(output_f, 'clasificacion.csv'))
# tabla['GLOSA_COMU'] = dic[tabla.index]
tabla.sort_values('sup_promedio')
```

```python
# Total de unidades por comuna y año
tabla = dfhr.groupby(['GLOSA_COMU', pd.DatetimeIndex(dfhr['FECHA_PE']).year])['CANTIDAD_U'].sum().unstack()
tabla
```

```python
# Promedio de pisos por región y año
tabla = dfhr.groupby(['GLOSA_COMU' , pd.DatetimeIndex(dfhr['FECHA_PE']).year])['NUM_PISOS'].mean().sort_values().unstack()
tabla
```

```python
# Máximo de pisos por región y año
tabla = dfhr.groupby(['GLOSA_COMU' , pd.DatetimeIndex(dfhr['FECHA_PE']).year])['NUM_PISOS'].max().sort_values().unstack()
tabla
```
